```python
with open("frankenstein.txt", encoding="utf-8") as f:
    frank = f.read()
    # litery
    from collections import Counter
    counts = Counter(frank)
    letters = sum( n for c, n in counts.items() if c.isalpha() )
    # cyfry
    digits = sum( n for c, n in counts.items() if c.isdigit() ) 
    # interkpunkcja
    import string
    punct = sum( n for c, n in counts.items() if c in string.punctuation) 
    print(f"3.2 W treści książki każdy z następujących symboli wystąpił w ilości:\nlitery: {letters},\ncyfry: {digits},\nznaki interpunkcyjne: {punct}")

print("3.3 Symbole w książce od najczęściej występującego:")
for char, count in counts.most_common():
    if char not in (" ", "\n", "\t", "\ufeff"): #ignorujemy spacje, tab, nowa linia, to nie są znaki intermpunkcyjne
        print(f"{repr(char)} : {count}")

#chars_list = [repr(c) for c in counts.keys()]
#print(chars_list)
```

```python
 # HOSPITALIZACJE OSÓB CHORYCH NA COVID 19
hosp = (15444, 16144, 16427, 17223, 18160, 18654, 19114)

def transform2(ts, func):
# ts   – sekwencja danych (np. liczby hospitalizacji
# func – funkcja, która przyjmuje dwa i zwraca wynik ich przekształcenia
    return [func(v0, v1) for v0, v1 in zip(hosp[:-1], hosp[1:])]
    # hosp[:-1] - wszystkie elementy oprócz ostatniego
    # hosp[1] - wszystkie elementy oprócz pierwszego
    # zip(...) łączy oba ciągi w pary: (dzień_i, dzień_i+1)
    # pętla for:
    #   v0 – wartość z dnia poprzedniego
    #   v1 – wartość z dnia następnego
    # func(v0, v1) wywołuje przekazaną funkcję (oblicza różnicę między dniami)
    # list comprehension tworzy listę wyników
transform2(
    hosp,                       # przekazujemy dane (krotkę hospitalizacji)
    lambda x0, x1: x1 - x0      # funkcja anonimowa obliczająca różnicę
)

print(f"Zmienna hosp to krotka z liczbą hospitalizacji w kolejnych dniach.\nW 7 linii wywołana jest anonimowa funkcja (lambda), która oblicza przyrost hospitalizacji między kolejnymi dniami (x0 dzień poprzeni, x1 dzień następny).\nW linii 5 zip(hosp[:-1], hosp[1:] tworzy pary (dzień poprzedni, dzień następny): (15444, 16144), (16144, 16427) itd.\nFunkcja [func(v0, v1) for v0, v1 in zip(...)] podstawia wartości dnia poprzedniego i następnego i zapisuje wynik do listy.\nFinalny wynik ma o jeden element mniej niż oryginalna krotka.")

percent_change = transform2(
    hosp,
    lambda x0, x1: (x1 - x0) / x0 * 100
)

print("\nDobowa zmiana procentowa osób hostpitalizowanych:")
for i, pct in enumerate(percent_change, start=1):
    print(f"Dzień {i} -> {i+1}: {pct:.2f}%")
```

```python
class Sample:
    # Konstruktor klasy
    def __init__(self, values):
        # values – lista obserwacji próbki (zapisujemy ją jako atrybut obiektu)
        self._values = list(values)  # kopia listy (bezpieczniej)

    # Zwraca wszystkie wartości próbki
    def get_vals(self):
        return self._values

    # Zwraca wartość jednej obserwacji (index – indeks elementu w próbce)
    def get_val(self, index):
        return self._values[index]

    # Ustawia nową wartość obserwacji (index – indeks elementu; value – nowa wartość)
    def set_val(self, index, value):
        self._values[index] = value

    # Dodaje nową obserwację na koniec próbki
    def add_val(self, value):
        self._values.append(value)

# Sprawdzenie poprawności działania

s = Sample([1, 2, 3])

print(s.get_vals()) # [1, 2, 3]

print(s.get_val(1)) # 2

s.set_val(1, 222)
print(s.get_vals()) # [1, 222, 3]

s.add_val(12345)
print(s.get_vals()) # [1, 222, 3, 12345]

```

```python
class ExtSample(Sample):
    # Oblicza sumę wartości w próbce
    def sum(self):
        return sum(self._values)

    # Oblicza iloczyn wartości w próbce
    def mul(self):
        result = 1
        for v in self._values:
            result *= v
        return result

    # Oblicza średnią arytmetyczną
    def avg(self):
        return sum(self._values) / len(self._values)

# Sprawdzenie poprawności działania
es = ExtSample([1, 2, 3, 4])

print(es.get_vals()) # [1, 2, 3, 4]

es.add_val(5)
print(es.get_vals()) # [1, 2, 3, 4, 5]

print(es.sum()) # 15

print(es.mul()) # 120

print(es.avg()) # 3.0
```

```python
class Vector:
    # Konstruktor klasy
    def __init__(self, values):
        # values – lista współrzędnych wektora
        self._values = list(values)
    # Zwraca współrzędne wektora
    def get_vals(self):
        return self._values

    # Przeciążenie operatora +
    # Dodawanie dwóch wektorów
    def __add__(self, other):
        # Sprawdzenie, czy wektory mają ten sam rozmiar
        if len(self._values) != len(other._values):
            raise ValueError("Wektory muszą mieć ten sam wymiar")
        # Dodawanie współrzędnych element po elemencie
        result = [
            v1 + v2
            for v1, v2 in zip(self._values, other._values)
        ]
        # Zwracamy nowy obiekt Vector
        return Vector(result)


    # Przeciążenie operatora *
    # Iloczyn skalarny dwóch wektorów
    def __mul__(self, other):
        # Sprawdzenie zgodności wymiarów
        if len(self._values) != len(other._values):
            raise ValueError("Wektory muszą mieć ten sam wymiar")

        # Iloczyn skalarny: suma iloczynów współrzędnych
        return sum(
            v1 * v2
            for v1, v2 in zip(self._values, other._values)
        )
    
# Sprawdzenie poprawności działania
v1 = Vector([1, 2, 1, 5])
v2 = Vector([2, 3, 1, 4])

v3 = v1 + v2
print(v3.get_vals()) # [3, 5, 2, 9]

print(v1 * v2) # 29
```

```python
import pyreadr
import pandas as pd
import numpy as np

bank_reg = pyreadr.read_r('bank_register.rds')
df = bank_reg[None] # wyciągamy DataFrame

# Rozdzielenie kolumny 'id' na 'client_id' i 'agreement_id'
df[['client_id', 'agreement_id']] = df['id'].str.split('_', expand=True)
df['client_id'] = df['client_id'].astype(int)
df['agreement_id'] = df['agreement_id'].astype(int)
df = df.drop(columns=['id'])  # Usunięcie oryginalnej kolumny 'id'

# Konwersja kolumny 'date' na typ daty
df['date'] = pd.to_datetime(df['date'], format='%b %d, %Y')

# Konwersja kolumny 'income' na typ float
df['income'] = (df['income']
    .str.replace(r"[$.]", "", regex=True)
    .str.replace(",", ".", regex=False)
    .astype(float)
)

# Rozdzielenie kolumny 'demographic' na 'sex', 'age', 'child'
df[['sex', 'age', 'child']] = df['demographic'].str.split(',', expand=True)

df['age'] = df['age'].astype(int)
df['child'] = df['child'].astype(int)

df = df.drop(columns=['demographic'])  # Usunięcie oryginalnej kolumny 'demographic'

# Rozdzielenie kolumny 'products' na kolumny: 'dep', 'cre', 'mor'
df['dep'] = df['products'].str.contains(r"\bDEP\b")
df['cre'] = df['products'].str.contains(r"\bCRE\b")
df['mor'] = df['products'].str.contains(r"\bMOR\b")

df = df.drop(columns=['products'])  # Usunięcie oryginalnej kolumny 'products'

# Ustawienie kolejności kolumn
df = df[['client_id', 'agreement_id', 'date', 'income', 'sex', 'age', 'child', 'dep', 'cre', 'mor']]
print(df.head())
```
